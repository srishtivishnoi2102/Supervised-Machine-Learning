## **Relative Frequency Distribution in Coin Toss Experiment**

**In one coin toss experiment, we will toss 'n' coins simultaneously and count the number of heads(0<=X<=n).
We will continue this experment for 'N' times and will note the number of heads observed in each experiment.
Then we will count the frequency of each observed value of X(number of heads) and then measure the relative frequency and will plot the distribution then**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

#N=int(input("Enter the number of times you want to perform the experiment with the dice : "))
#n=int(input("Enter the number of times dice will be thrown in 1 experiment : "))

def plot_distribution(N,n):
  data = np.random.randint(low = 0, high = 2, size = (N,n)) 
  counts=list(np.count_nonzero(data,axis=1))
  possible_count=list(set(counts))
  count_relfreq=list(map(lambda val:counts.count(val)/N, possible_count))
  plt.bar(possible_count, count_relfreq)
  plt.plot(possible_count, count_relfreq)
  plt.xlabel("Possible Number of Heads in 1 experiment: X")
  plt.ylabel("Relative Frequency : RV(X)")
  title="Relative Frequency Distribution of Number of Heads: when "+str(n)+" coins are tossed "+str(N) +" times"
  plt.title(title)
  plt.show()
```

**In the below plot,N is very small, so relative frequency will fluctuate a lot, and plot will change everytime we repeat the setup (try running this same cell multiple times)**

```python
plot_distribution(50,10)
```

As we increase the value of N, i.e., Number of experiments, we can see our fluctuation in the distribution will reduce, i.e., the distribution wont fluctuate much if we run the cell multiple times

```python
plot_distribution(1000,10)
```

**As the value of N is getting larger and limiting toward population size, the relative frequency values are limiting to a stable value, i.e., probability and our relative frequency distribution is approaching the Probability Distribution**

```python
plot_distribution(100000,10)
```
