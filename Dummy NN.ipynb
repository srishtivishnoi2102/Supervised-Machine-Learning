```python
M= 10     # number of rows in dataset
N= 3     # number of columns in dataset

num_of_neuron_J=4      # number of neuron in layer J
num_of_neuron_K=3      # number of neuron in layer K
num_of_neuron_L=2       # number of neuron in layer L

num_of_neuron_in_output=1  
```

```python
input_data=np.random.randn(M,N)
print("input_data.shape: ",input_data.shape)
input_data
```

## weight matrix between layer I and J , with shape (num_of_neurons in I,num_of_neurons in J)

```python
weights_ij=np.random.randn(N,num_of_neuron_J)
print("weights_ij.shape :",weights_ij.shape)
weights_ij
```

# J= I.Weights(ij)

```python
layer_J= np.dot(input_data,weights_ij)
print("layer_J.shape : ",layer_J.shape)
layer_J
```

## weight matrix between layer J and K , with shape (num_of_neurons in J,num_of_neurons in K)

```python
weights_jk= np.random.randn(num_of_neuron_J,num_of_neuron_K)
print("weights_jk.shape: ",weights_jk.shape)
weights_jk
```

# K= J.Weights(jk)


```python
layer_K= np.dot(layer_J,weights_jk)
print("layer_K.shape : ",layer_K.shape)
layer_K
```

# weight matrix between layer K and L , with shape (num_of_neurons in K,num_of_neurons in L)


```python
weights_kl= np.random.randn(num_of_neuron_K,num_of_neuron_L)
print("weights_kl.shape: ",weights_kl.shape)
weights_kl
```

# L= K.Weights(kl)


```python
layer_L= np.dot(layer_K,weights_kl)
print("layer_L.shape : ",layer_L.shape)
layer_L
```

# weight matrix between layer K and L , with shape (num_of_neurons in K,num_of_neurons in L)


```python

weights_lo= np.random.randn(num_of_neuron_L,num_of_neuron_in_output)
print("weights_lo.shape: ",weights_lo.shape)
weights_lo
```

# output = L.weights(lo)

```python
output_layer=np.dot(layer_L,weights_lo)
print("output_layer.shape : ",output_layer.shape)
output_layer
```
